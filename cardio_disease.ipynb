LOGISTIC REGRESSION: baseline model

```python
X = df[['age','sex','cp','trestbps','chol','fbs','restecg','thalach',
        'exang','oldpeak','slope','ca','thal']]

y = df['target']

# Convert to binary target: 0 = no disease, 1 = disease
y_binary = y.apply(lambda val: 0 if val == 0 else 1)

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score,f1_score

# Use y_binary instead of y
X_train, X_test, y_train, y_test = train_test_split(
    X, y_binary, test_size=0.2, random_state=42, stratify=y_binary
)

# Train Logistic Regression
model = LogisticRegression()
model.fit(X_train, y_train)

# Predictions
predict = model.predict(X_test)

# Evaluation
print("Accuracy:", accuracy_score(y_test, predict))
# print("\nClassification Report:\n", classification_report(y_test, predict))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, predict))
print("\nF1 Score:", f1_score(y_test, predict))
```

```python
train_score = model.score(X_train, y_train)
test_score = model.score(X_test, y_test)

print(f"Training Accuracy: {train_score:.4f}")
print(f"Test Accuracy: {test_score:.4f}")
```

```python
train_f1=f1_score(y_train, model.predict(X_train))
test_f1=f1_score(y_test, model.predict(X_test))

print(f"Training F1 Score: {train_f1:.4f}")
print(f"Test F1 Score: {test_f1:.4f}")
```

ROC = Receiver Operating Characteristic Curve
It’s a graph that shows how well your classification model separates the two classes —
for you: “Cardio Disease” (1) vs “No Disease” (0).

It plots:

True Positive Rate (TPR) → also called Recall
False Positive Rate (FPR) → how many non-disease people were wrongly predicted as diseased
thresholds → These are cutoff probability values used to decide whether a person is predicted “diseased (1)” or “not (0)”. (threshold of 0.5 - 1 else - 0)

AUC = Area Under the ROC Curve

It’s a single number summary of the ROC curve.

It tells how good your model is at distinguishing between classes.

```python
from sklearn.metrics import roc_auc_score,roc_curve

fpr, tpr, thresholds = roc_curve(y_test, predictions)
auc=roc_auc_score(y_test,predictions)

plt.plot(fpr, tpr, label=f'ROC Curve (AUC = {auc:.2f})')
plt.plot([0,1], [0,1], 'k--', label='Random Guessing')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve for Cardiovascular Disease Prediction')
plt.legend()
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score,f1_score

X_train, X_test, y_train, y_test = train_test_split(X, y_binary, test_size=0.2, random_state=42)
X_train.dropna(inplace=True)
X_test.dropna(inplace=True)
y_train=y_train[X_train.index]
y_test=y_test[X_test.index]


model = KNeighborsClassifier(n_neighbors=5,p=2,metric='minkowski')
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
print("\nAccuracy Score:")
print(accuracy_score(y_test, y_pred))
print("\nF1 Score:")
print(f1_score(y_test, y_pred))
```

```python
from ucimlrepo import fetch_ucirepo 
  
# fetch dataset 
heart_disease = fetch_ucirepo(id=45) 
  
# data (as pandas dataframes) 
X = heart_disease.data.features 
y = heart_disease.data.targets 
  
# # metadata 
# print(heart_disease.metadata) 
  
# # variable information 
# print(heart_disease.variables) 
print(X.head(20))
print(X.shape)
```

```python
df=X.copy()
df['target']=y

df=df.fillna(df.mean(numeric_only=True))

X =df[['age','cp','trestbps','chol','fbs','exang','oldpeak','slope','ca','thal']]
y=df['target']

y_binary=y.apply(lambda val:0 if val==0 else 1)
```

```python
df.shape
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.callbacks import EarlyStopping
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.layers import ReLU,PReLU,LeakyReLU
from tensorflow.keras.optimizers import Adam
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score,f1_score
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y_binary,test_size=0.2,random_state=0,stratify=y_binary)
```

```python
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
classifier=Sequential()

#input layer
classifier.add(Dense(units=32,activation='relu'))

#first hidden layer
classifier.add(Dense(units=24,activation='relu'))

#second hidden layer
classifier.add(Dense(units=16,activation='relu'))

#third hidden layer
classifier.add(Dense(units=8,activation='relu'))
classifier.add(Dropout(0.2))

#output layer
classifier.add(Dense(units=1,activation='sigmoid'))  #1 neuron for binary classification
```

```python
earlystop=EarlyStopping(monitor='val_loss',patience=10,restore_best_weights=True)
```

```python
import tensorflow
adam=tensorflow.keras.optimizers.Adam(learning_rate=0.001)

classifier.compile(optimizer=adam,loss='binary_crossentropy',metrics=['accuracy'])
```

```python
model=classifier.fit(X_train,y_train,validation_split=0.33,batch_size=16,epochs=50,callbacks=[earlystop])
```

```python
y_pred = classifier.predict(X_test)
import numpy as np
y_pred = (y_pred > 0.5).astype(int)
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Confusion Matrix:", confusion_matrix(y_test, y_pred))
```

```python
train_acc = model.history['accuracy'][-1]
train_loss = model.history['loss'][-1]

loss, accuracy = classifier.evaluate(X_test, y_test)

print(f"Train Loss: {train_loss}")
print(f"Test Loss: {loss}")
print(f"Training Accuracy: {train_acc}")
print(f"Test Accuracy: {accuracy}")
```
